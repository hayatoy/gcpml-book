## Chapter3.4 Cloud Natural Language API

[API Key](https://console.cloud.google.com/apis/credentials)を入力してEnter

```python
import getpass
APIKEY = getpass.getpass()
```

```python
# クライアントライブラリのロードとインスタンスの作成
from googleapiclient.discovery import build
nl_service = build('language', 'v1', developerKey=APIKEY)
```

```python
# 解析する文章
text = u'六本木ヒルズにあるGoogleのオフィスには山手線駅名の会議室があります。'

# エンティティ分析を実行
response = nl_service.documents().analyzeEntities(
  body={
    'document': {
      'type': 'PLAIN_TEXT',
      'content': text,
      'language': 'ja',
    },
    'encodingType': 'UTF8',
  }
).execute()


# 結果の表示
for entity in response['entities']:
    print('=' * 20)
    print(u'　　　単語: {}'.format(entity['name']))
    print(u'　　　種類: {}'.format(entity['type']))
    print(u'メタデータ: {}'.format(entity['metadata']))
    print(u'　　重要度: {}'.format(entity['salience']))
```

```python
# 解析する文章
text = u"""
昨日食べたフレンチはとても美味しかったです。
でも値段が高すぎるのでもう行かないでしょう。"""

# 感情分析を実行
response = nl_service.documents().analyzeSentiment(
  body={
    'document': {
      'type': 'PLAIN_TEXT',
      'content': text,
      'language': 'ja',
    },
    'encodingType': 'UTF8',
  }
).execute()

print(u'全文の感情値: {}, {}'.format(response['documentSentiment']['score'],
                                   response['documentSentiment']['magnitude']))
print('')

print(u'文章毎の感情値;')
for s in response['sentences']:
    print('=' * 20)
    print(u'文章: {}'.format(s['text']['content']))
    print(u'感情値: {}, {}'.format(s['sentiment']['score'],
                                  s['sentiment']['magnitude']))
```
