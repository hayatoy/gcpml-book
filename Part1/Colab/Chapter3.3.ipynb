## Chapter3.3 Cloud Translation API

[API Key](https://console.cloud.google.com/apis/credentials)を入力してEnter

```python
import getpass
APIKEY = getpass.getpass()
```

```python
# クライアントライブラリのロードとインスタンスの作成
from googleapiclient.discovery import build
translate_service = build('translate', 'v2', developerKey=APIKEY)
```

```python
# 翻訳したい文章
text = u'GCPで機械学習が簡単に学べました。'
# 翻訳先言語
target = 'en'

translation = translate_service.translations().list(
  target=target,
  q=text,
  model='nmt').execute()

print(u'翻訳元: {}'.format(text))
print(u'翻訳後: {}'.format(translation['translations'][0]['translatedText']))
```

```python
wrong_translation = translate_service.translations().list(
    q=u'手紙',
    target='en').execute()
wrong_translation['translations'][0]['translatedText']
```

```python
right_translation = translate_service.translations().list(
    q=u'手紙',
    source='zh',
    target='en').execute()
right_translation['translations'][0]['translatedText']
```

```python
results = translate_service.languages().list().execute()

print(u'対応言語数:{}'.format(len(results['languages'])))

for language in results['languages']:
    print(u'{},'.format(language['language']))
```
